```python
import math 

x = []
y = []
t = []
f = 5885000
dryMass = 31200 
fuelMass = 474646
totalMass = dryMass + fuelMass



def findAcceleration(f, m):
    return(f/m)

i = 0
while fuelMass > 0: 
    x.append(findAcceleration(f,totalMass))
    y.append(totalMass)
    t.append(i)
    fuelMass -= 100
    totalMass = dryMass + fuelMass
    i+= 1
```

```python
import math
import numpy as np 
from matplotlib import pyplot as plt 


plt.title("Falcon 9 Acceleration Stage 1 Curve") 
plt.xlabel("Acceleration") 
plt.ylabel("Mass") 
plt.plot(x,y) 
plt.show()
```

```python
plt.title("Falcon 9 Acceleration Stage 1 Curve") 
plt.xlabel("Time") 
plt.ylabel("Acceleration") 
plt.plot(t,x) 
plt.show()
```
